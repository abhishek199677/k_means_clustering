```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs   # to generate blobs dataset
from sklearn.cluster import KMeans       #  k-means clustering
from sklearn.metrics import silhouette_score    # silhouette score for clustering
```

```python
x,y = make_blobs(n_samples=300 , centers=4 , cluster_std=.60 , random_state=30) 
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:,0] ,x[:,1])  # selects all the rows of first column and second column
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++',random_state=30)
```

```python
y_kmeans =  kmeans.fit_predict(x)     # fitting the model and predicting
```

```python
cent = kmeans.cluster_centers_    # centroids of the clusters
```

```python
plt.scatter(x[:,0] , x[:,1],c=y_kmeans)     # plotting the clusters
plt.scatter(cent[:,0] , cent[:,1] , c = 'red')   # plotting the centroids
```

```python
# elbow method to find the optimal number of clusters

inertia = []
k = range(1,10)
for i in k :
    km = KMeans(n_clusters=i,init='k-means++')
    km.fit(x)
    inertia.append(km.inertia_)
```

```python
inertia
```

```python
plt.plot(k,inertia,marker='o')
```

```python
silhouette_score(x,y_kmeans)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
irish = load_iris()
```

```python
x = irish.data
y = irish.target
```

```python
x
```

```python
y 
```

```python
x_train,x_test , y_train,y_test = train_test_split(x,y,test_size=.20)
```

```python
nbmodel = GaussianNB()
nbmodel.fit(x_train,y_train)
```

```python
y_pred = nbmodel.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print("classification report \n",classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
